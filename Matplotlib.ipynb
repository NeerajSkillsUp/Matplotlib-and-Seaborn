Let's plot the sine wave

```python
import numpy as np
x = np.arange(0, 4*m.pi,0.1)
y = np.sin(x)

z=sns.lineplot(x=x,y=y)
z.set_title("Sine wave form")
```

sine and cosine curves

```python
x = np.arange(0, 5 * m.pi, 0.1)

y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1,and

# Set the first such subplot as active.
plt.subplot(2, 2, 1)
# Make the first plot
z=sns.lineplot(x=x, y=y_sin)
z.set_title('Sine')

# Set the second subplot as active.
plt.subplot(2, 2, 4)                 # positions 2 and 3 still empty on 2x2 grid
# Make the second plot
z=sns.lineplot(x=x,y = y_cos)
z.set_title('Cosine')
```

boxplot

```python
import numpy as np
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
sns.boxplot(data)
```


The code data = [np.random.normal(0, std, 100) for std in range(1, 4)] creates a list of 3 NumPy arrays, each of which contains 100 random numbers drawn from a normal distribution with a mean of 0 and a standard deviation of 1, 2, or 3. The code first creates a range of values from 1 to 3, which will be used as the standard deviations for the normal distributions. Then, it uses a for loop to iterate over the range and generate a NumPy array of 100 random numbers for each standard deviation. Finally, it creates a list of the 3 NumPy arrays.

TimeSeriesDataset - real time data

```python
import pandas as pd
series = pd.read_csv('timeSeriesDataset.csv',header=0,index_col=0,parse_dates=True).squeeze("columns")
print(series.head())
df = pd.DataFrame(series)
sns.lineplot(df)
```

Minimum Daily Temperature Yearly Line Plots

```python
groups = series.groupby(pd.Grouper(freq='YE'))         # YE means annual
years = pd.DataFrame()
for name, group in groups:
    years[name.year] = group.values
years.plot(subplots=True,legend=False)
```

Minimum Daily Temperature Yearly Histogram

```python
sns.displot(df)
```

kdeplot (kernel density estimate)

```python
sns.kdeplot(df)
```

Minimum Daily Temperature Yearly Box and Whisker Plots

```python
groups = series.groupby(pd.Grouper(freq='YE'))
years = pd.DataFrame()
for name, group in groups:
    years[name.year] = group.values
years.boxplot()
```

Minimum Daily Temperature Yearly Heat Map Plot

```python
import matplotlib.pyplot as plt
groups = series.groupby(pd.Grouper(freq='YE'))
years = pd.DataFrame()
for name, group in groups:
    years[name.year] = group.values
years = years.T
plt.matshow(years, interpolation=None, aspect='auto')
```

Minimum Daily Temperature Lag Plot

```python
from pandas.plotting import lag_plot
lag_plot(series)
```

Minimum Daily Temperature Autocorrelation Plot

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(series)
```
